## Predictive performance for all models in IDF for March 2025

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the master metrics file for the region you're analyzing
df = pd.read_csv("Predictions/IDF/metrics_master_idf.csv") 

# Optional : make sure "Model" column is treated as a categorical with ordered model names
model_order = ["M12", "M18", "M24", "M36", "M48", "ALL_MODELS"]
df["Model"] = pd.Categorical(df["Model"], categories=model_order, ordered=True)

# Create the boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="Model", y="R2", palette="viridis", showmeans=True)

plt.title("Distribution of R2 Scores per Model (IDF March 2025)")
plt.xlabel("Model")
plt.ylabel("R2 SCore")
plt.grid(True, linestyle="--", alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the master metrics file for the region you're analyzing
df = pd.read_csv("Predictions/IDF/metrics_master_idf.csv") 

# Optional : make sure "Model" column is treated as a categorical with ordered model names
model_order = ["M12", "M18", "M24", "M36", "M48", "ALL_MODELS"]
df["Model"] = pd.Categorical(df["Model"], categories=model_order, ordered=True)

# Create the boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="Model", y="R2", palette="viridis", showmeans=True, showfliers=False)

plt.title("Distribution of R2 Scores per Model (IDF March 2025)")
plt.xlabel("Model")
plt.ylabel("R2 SCore")
plt.grid(True, linestyle="--", alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming df is your master metrics DataFrame
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="Model", y="R2", showmeans=True, showfliers=False, palette="coolwarm")

# Calculate means and variances for annotation
model_stats = df.groupby("Model")["R2"].agg(["mean", "std"]).reset_index()

# Annotate each box
for i, row in model_stats.iterrows():
    plt.text(i, -0.25, f"μ={row['mean']:.2f}\nσ={row['std']:.2f}",
             ha="center", va="bottom", fontsize=10, fontweight="bold")

plt.title("R² Distribution by Model (Mean & Variance Displayed) (IDF)")
plt.ylabel("R² Score")
plt.grid(True, linestyle="--", alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming df is your master metrics DataFrame
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="Model", y="MAE", showmeans=True, showfliers=False, palette="coolwarm")

# Calculate means and variances for annotation
model_stats = df.groupby("Model")["MAE"].agg(["mean", "std"]).reset_index()

# Annotate each box
for i, row in model_stats.iterrows():
    plt.text(i, 500.25, f"μ={row['mean']:.2f}\nσ={row['std']:.2f}",
             ha="center", va="bottom", fontsize=10, fontweight="bold")

plt.title("MAE Distribution by Model (Mean & Variance Displayed) (IDF)")
plt.ylabel("MAE Score")
plt.grid(True, linestyle="--", alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Assuming df is your master metrics DataFrame
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="Model", y="RMSE", showmeans=True, showfliers=False, palette="coolwarm")

# Calculate means and variances for annotation
model_stats = df.groupby("Model")["RMSE"].agg(["mean", "std"]).reset_index()

# Annotate each box
for i, row in model_stats.iterrows():
    plt.text(i, 500.25, f"μ={row['mean']:.2f}\nσ={row['std']:.2f}",
             ha="center", va="bottom", fontsize=10, fontweight="bold")

plt.title("RMSE Distribution by Model (Mean & Variance Displayed) (IDF)")
plt.ylabel("RMSE Score")
plt.grid(True, linestyle="--", alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x="Run_time", y="R2", hue="Model", palette="Set2", showfliers=False)
plt.title("R² Score Distribution by Run Time and Model")
plt.xlabel("Run Time")
plt.ylabel("R² Score")
plt.legend(title="Model")
plt.grid(True, linestyle="--", alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
mean_r2 = df.groupby("Model")["R2"].mean().sort_values(ascending=False)
mean_r2.plot(kind="bar", figsize=(10, 5), title="Average R² per Model", color="steelblue")
plt.ylabel("Average R²")
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
corr = df[["MAE", "RMSE", "R2", "MAE / Mean", "RMSE / Mean"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Between Evaluation Metrics")
plt.tight_layout()
plt.show()
```
